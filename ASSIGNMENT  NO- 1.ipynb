```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=pd.read_excel(r"C:\Users\SANDRA\Downloads\data.xls")
```

```python
iris["Classification"].value_counts()
```

```python
iris.plot(kind="scatter",x="PL", y= "PW").set(title="SCATTER PLOT- PL vs PW")
```

```python
sns.jointplot(x="SL",y="PL", data=iris,color="g",hue="Classification")     ###joint plot of SL vs PL,scatterplot and histogram combined together,iris versicolor and iris verginica are little similar bus iris setosa is different from others
```

```python
sns.boxplot(x="Classification", y="PL",data=iris).set(title=" BOX PLOT")      # analysis of 3 species with respect to PL,also te median value of data of each species
```

```python
sns.pairplot(iris, hue="Classification", size=3).add_legend().set(title=" PAIR PLOT")
```

```python
sns.lineplot(x="SL",y="PL", data=iris,hue="Classification").set(title="SL vs PL-Line plot")
```

```python
sns.lineplot(x="Classification",y="SW", data=iris, marker="*",markerfacecolor="r",markersize=20).set(title="variation of SW of each species") 
```
